```python
import pandas as pd
from datasets import load_dataset

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country']
```

```python
df['job_country'].value_counts()
```

```python
df['job_country'].isin(['India']).any() # help us find a value in the column
```

```python
ind_jobs = df[df['job_country'] == 'India']
```

```python
ind_jobs
```

```python
ind_jobs = ind_jobs[ind_jobs['salary_year_avg'].notna()]
```

```python
ind_jobs
```

```python
ind_jobs.groupby('job_title_short').count()
```

```python
ind_jobs.groupby('job_title_short')['salary_year_avg'].min()
```

```python
ind_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count'])
```

```python
ind_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median') # sort using median in the groupby
```

```python
ind_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```
