```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleaning data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

# Filter for US Data Analyst roles 

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame() # to_frame: converts it to a data frame

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
plt.title('Counts of Jobs Posting Location for Data Analyst in US')
plt.ylabel('')
plt.xlabel('Number of Jobs')
plt.show()
```

```python
# making a dict which we can use in a loop since we don't wanna go and write the below again and again.

dict_column = {
    'job_work_from_home' : 'Work From Home',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12,3))

for i, (column, title) in enumerate(dict_column.items()):  # we are running a loop in the items provided in the dict. 
    ax[i].pie(df_DA_US[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True']) # labels to show what each section means.
    ax[i].set_title(title)

plt.show()
```

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame() # to_frame: converts it to a data frame

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
plt.title('Counts of Company names hiring for Data Analyst jobs in US')
plt.ylabel('')
plt.xlabel('Number of Jobs')
plt.show()
```
