# Exercise - Investigate Trend Skills

```python
# Importing libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA
```

```python
# we need to aggregate the skills in monthly basis. So we can check the trending skills for that month's data analyst job.

df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')

df_DA_explode
```

```python
# we can see it repeating now we need to take a count of it.

df_DA_explode = df_DA.explode('job_skills')

# to get the required data we need to do a pivot.

df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size')
```

```python
# We can see non-values present in the data, so we need to use a parameter in the pivot table.

df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
# to get a data which we can plot, since there is too many skills even after seeing data defined as 0 it is gettting plotted in the graph. We should too a total of all the skills and add the row at the end.

df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot
```

```python
# now we need to collate the total and sort values to see the top skills.

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index] # we are assigning this under the df_DA_pivot so it can sort it as per this condition we have formed, the top ones at front and least ones as the right hand corner.

df_DA_pivot.drop('Total', inplace=True) # since this not really needed now hence removing.
```

```python
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

df_DA_pivot

df_DA_pivot.iloc[:, :5].plot(kind='line') # will iloc since we don't wanna callout each rows, specifically require from 0 to 5.

plt.title('Top 5 skills for Data Analyst based on month')
plt.ylabel('Count')
plt.xlabel('')
plt.legend().set_visible(True)
plt.show()
```
