```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

df
```

```python
months = df['job_posted_month'].unique()

# Creating a dict for months.

dict_months = {month: df[df['job_posted_month'] == month] for month in months}
```

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True) # had to put dict_months under another bracket since only one args can be defined under the concat.
```

```python
df_q1['job_posted_month'].value_counts() # to check how many values are there after the concat is done and we get a new data.
```
